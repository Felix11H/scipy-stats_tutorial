# scipy.stats


```python
%matplotlib inline
import numpy as np
import pylab as pl
```

```python
from scipy.stats import norm

mean = 5
std = 1
```

```python
sample = norm.rvs(mean,std, size=10000)
```

```python
out = pl.hist(sample, bins=25, normed=True)
```

```python
xs = np.arange(1,9,0.01)
fxs = norm.pdf(xs, mean, std)
pl.plot(xs,fxs, 'r', linewidth=5)
```

```python
out = pl.hist(sample, bins=25, normed=True)
pl.plot(xs,fxs, 'r', linewidth=5)
```

# custom variables with rv_continuous

```python
from scipy.stats import rv_continuous

class Custom_rv(rv_continuous):

    def __init__(self, h, lower_bound, upper_bound):

        self.h = h

        rv_continuous.__init__(self, a=lower_bound, b=upper_bound)


    def _pdf(self, x):
        if -1.75 < x <-1.25:
            return 0.2/1.1
        elif -1 < x < 0:
            return self.h*(x+1)/1.1
        elif 0 <= x < 1:
            return -1*self.h*(x-1)/1.1
        else:
            return 0
```

```python
RV = Custom_rv(1.,-2.5,2)
```

```python
xs = np.arange(-2.5,2,0.01)
ys = [RV.pdf(x) for x in xs]
```

```python
pl.plot(xs,ys, 'r', linewidth=5)
```

```python
samples = RV.rvs(size=5000)
out = pl.hist(samples, bins=30, normed=True)
```

```python
out = pl.hist(samples, bins=30, normed=True)
pl.plot(xs,ys, 'r', linewidth=5)
```
